```python
#Quando você instala o chromedriver, ele precisa ficar guardado na mesma pasta que o Python está.
#Depois de intalar o selenium, basta olhar na primeira linha a localização do Python no seu computador.
#Neste caso, por exemplo, localização: c:\users\jjosi\anaconda3
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
#As linhas de código a seguir são para caso queira que as automações rodem em segundo plano, sem o chrome aparecer.
#from selenium.webdriver.chrome.options import Options
#chrome_options = Options()
#chrome_options.headless = True
#navegador = webdriver.Chrome(options = chrome_options)
navegador = webdriver.Chrome()
navegador.get("https://www.google.com.br/?hl=pt-BR")
#As aspas no código a seguir são aspas simples. 
#Para identificar o xpath dentro do site:
#Clicar com o botão direito do mouse, inspecionar, clicar na setinha que está no canto superior esquerdo da pasta que apareceu
#Selecionar com a seta a região, levar o cursor onde está azul, botão direito, copy, copy xpath.
navegador.find_element('xpath','//*[@id="APjFqb"]').send_keys("Cotação Dólar")
#Caso quisesse apenas clicar na região o código seria navegador.find_element('xpath'/'//*[@id="APjFqb"]').click
#Mas neste caso queremos escrever na região, por isso fica .send_keys
#Para clicar em alguma tecla do teclado, precisamos do common.keys
navegador.find_element('xpath','//*[@id="APjFqb"]').send_keys(Keys.ENTER)
#Pegando apenas o número da região selecionada.
navegador.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
cotacao_dolar = navegador.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)
navegador.get("https://www.google.com.br/?hl=pt-BR")
navegador.find_element('xpath','//*[@id="APjFqb"]').send_keys("Cotação Euro")
navegador.find_element('xpath','//*[@id="APjFqb"]').send_keys(Keys.ENTER)
navegador.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
cotacão_euro = navegador.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacão_euro)
navegador.get("https://www.melhorcambio.com/ouro-hoje")
navegador.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
cotacão_ouro = navegador.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
print(cotacão_ouro)
```

```python
#Se o número tiver casa decimal, o python só consegue fazer a conta com esse número de tiver um ponto em vez da vírgula.
#Vamos resolver isso no caso da cotação do ouro.
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
navegador = webdriver.Chrome()
navegador.get("https://www.melhorcambio.com/ouro-hoje")
navegador.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
cotacão_ouro = navegador.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
cotacão_ouro = cotacão_ouro.replace(",",".")
print(cotacão_ouro)
#Caso queira que o navegador feche depois que o selenium executar o código:
#navegador.quit()
```

```python
import pandas as pd
tabela = pd.read_excel(r"C:\Users\jjosi\Downloads\Base Atualizada 2.xlsx")
tabela.drop(columns = ['Unnamed: 7','Unnamed: 8','Unnamed: 9'], inplace = True)
display(tabela)
```

```python
#Alterando a coluna cotação
#Cotação dólar
#O código a seguir quer dizer que sempre que a coluna moeda for igual ao texto dólar, a coluna cotação vai ser igual a cotacão_dolar.
#Adicionei float antes de cotacão_dolar só para garantir que o formato vai ser em número. 
tabela["Moeda"] == "Dólar"
tabela.loc[tabela["Moeda"] == "Dólar","Cotação"] = float(cotacao_dolar)
#Alterando a cotação do euro
tabela["Moeda"] == "Euro"
tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(cotacão_euro)
#Alterando a cotação do ouro
tabela["Moeda"] == "Ouro"
tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(cotacão_ouro)
display(tabela)
```

```python
#Como a coluna preço de compra é a coluna preço original vezes a coluna cotação, precisamos atualiza a coluna preço de compra
tabela["Preço de compra"] = tabela["Preço Original"]*tabela["Cotação"]
#A coluna preço de venda é a coluna preço de compra vezes a coluna margem, também precisamos atualizar a coluna preço de venda
tabela["Preço de Venda"] = tabela["Preço de compra"]*tabela["Margem"]
display(tabela)
```

```python
#Agora, com a tabela atualizada, devemos exportá-la de volta para o excel.
#A formtação é melhor fazer no excel, pode exportar assim mesmo.
#Se colocar o mesmo nome do arquivo do excel, o python vai modificar o arquivo original, caso não queira que isso aconteça, 
#é só mudar o nome. Neste caso, vou mudar o nome.
tabela.to_excel("Base Atualizada 2 (1).xlsx")
#Para encontrar no seu computador essa nova tabela, é só pegar o nome dela, neste caso "Base Atualizada 2 (1)", abrir arquivos
#e jogar em "pesquisar em início".
```
